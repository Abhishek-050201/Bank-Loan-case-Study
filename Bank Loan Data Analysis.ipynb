# Income Bins

```python
def inc_bin(AMT_INCOME_TOTAL):
    if AMT_INCOME_TOTAL <=25000:
        return '0-25k'
    elif AMT_INCOME_TOTAL>25000 and AMT_INCOME_TOTAL<50000:
        return '25k-50k'
    elif AMT_INCOME_TOTAL>50000 and AMT_INCOME_TOTAL<75000:
        return '50k-75k'
    elif AMT_INCOME_TOTAL>75000 and AMT_INCOME_TOTAL<100000:
        return '75k-100k'
    elif AMT_INCOME_TOTAL>100000 and AMT_INCOME_TOTAL<125000:
        return '100k-125k'
    elif AMT_INCOME_TOTAL>125000 and AMT_INCOME_TOTAL<150000:
        return '125k-150k'
    elif AMT_INCOME_TOTAL>150000 and AMT_INCOME_TOTAL<175000:
        return '150k-175k'
    elif AMT_INCOME_TOTAL>175000 and AMT_INCOME_TOTAL<200000:
        return '175k-200k'
    elif AMT_INCOME_TOTAL>200000 and AMT_INCOME_TOTAL<225000:
        return '200k-225k'
    elif AMT_INCOME_TOTAL>225000 and AMT_INCOME_TOTAL<250000:
        return '225k-250k'
    elif AMT_INCOME_TOTAL>250000 and AMT_INCOME_TOTAL<275000:
        return '250k-275k'
    elif AMT_INCOME_TOTAL>275000 and AMT_INCOME_TOTAL<300000:
        return '275k-300k'
    elif AMT_INCOME_TOTAL>300000 and AMT_INCOME_TOTAL<325000:
        return '300k-325k'
    elif AMT_INCOME_TOTAL>325000 and AMT_INCOME_TOTAL<350000:
        return '325k-350k'
    elif AMT_INCOME_TOTAL>350000 and AMT_INCOME_TOTAL<375000:
        return '350k-375k'
    elif AMT_INCOME_TOTAL>375000 and AMT_INCOME_TOTAL<400000:
        return '375k-400k'
    elif AMT_INCOME_TOTAL>400000 and AMT_INCOME_TOTAL<425000:
        return '400k-425k'
    elif AMT_INCOME_TOTAL>425000 and AMT_INCOME_TOTAL<450000:
        return '425k-450k'
    elif AMT_INCOME_TOTAL>450000 and AMT_INCOME_TOTAL<475000:
        return '450k-475k'
    elif AMT_INCOME_TOTAL>475000 and AMT_INCOME_TOTAL<500000:
        return '475k-500k'
    else:
        return 'Above 5 Lakhs'
```

```python
df['Income Bins'] =  df['AMT_INCOME_TOTAL'].apply(inc_bin)
```

# Credit Bins

```python
def crt_bin(AMT_CREDIT):
    if AMT_CREDIT <=150000:
        return '0L-1.5L'
    elif AMT_CREDIT > 150000 and AMT_CREDIT<=200000:
        return '1.5L-2.0L'
    elif AMT_CREDIT >200000 and AMT_CREDIT <=250000:
        return '2.0L-2.5L'
    elif AMT_CREDIT > 250000 and AMT_CREDIT<=300000:
        return '2.5L-3.0L'
    elif AMT_CREDIT >300000 and AMT_CREDIT <=350000:
        return '3.0L-3.5L'
    elif AMT_CREDIT > 350000 and AMT_CREDIT<= 400000:
        return '3.5L-4.0L'
    elif AMT_CREDIT >400000 and AMT_CREDIT <= 450000:
        return '4.0L-4.5L'
    elif AMT_CREDIT > 450000 and AMT_CREDIT<= 500000:
        return '4.5L-5.0L'
    elif AMT_CREDIT >500000 and AMT_CREDIT <= 550000:
        return '5.0L-5.5L'
    elif AMT_CREDIT > 550000 and AMT_CREDIT<= 600000:
        return '5.5L-6.0L'
    elif AMT_CREDIT > 600000 and AMT_CREDIT <= 650000:
        return '6.0L-6.5L'
    elif AMT_CREDIT > 650000 and AMT_CREDIT<= 700000:
        return '6.5L-7.0L'
    elif AMT_CREDIT >700000 and AMT_CREDIT <=750000:
        return '7.0L-7.5L'
    elif AMT_CREDIT > 750000 and AMT_CREDIT<= 800000:
        return '7.5L-8.0L'
    elif AMT_CREDIT >800000 and AMT_CREDIT <=850000:
        return '8.0L-8.5L'
    elif AMT_CREDIT > 850000 and AMT_CREDIT<= 900000:
        return '8.5L-9.0L'
    else:
        return 'Above 9 Lakhs'
```

```python
df['Credit_Bins'] = df['AMT_CREDIT'].apply(crt_bin)
```
